```python
def risk_category(row):
    if row["credit_band"] == "Poor" and row["employment_type"] == "Unemployed":
        return "High Risk"
    elif row["credit_band"] == "Poor":
        return "High Risk"
    elif row["credit_band"] == "Fair" and row["debt_to_income_ratio"] > 0.6:
        return "Medium Risk"
    elif row["previous_defaults"] > 0:
        return "Medium Risk"
    else:
        return "Low Risk"

df["risk_category"] = df.apply(risk_category, axis=1)

df["risk_category"].value_counts(normalize=True) * 100

```

```python
df.groupby("risk_category")["default_flag"].mean() * 100
```
